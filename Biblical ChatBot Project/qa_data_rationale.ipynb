```python
import torch
import chromadb
from chromadb.config import Settings
from transformers import pipeline
from langchain import PromptTemplate, LLMChain
from langchain.llms import HuggingFacePipeline
```

```python
generate_text = pipeline(model="databricks/dolly-v2-3b", torch_dtype=torch.bfloat16,
                         trust_remote_code=True, device_map="auto", return_full_text=True)
```

```python
# template for an instrution with no input
prompt = PromptTemplate(
    input_variables=["instruction"],
    template="{instruction}")

# template for an instruction with input
prompt_with_context = PromptTemplate(
    input_variables=["instruction", "context"],
    template="{instruction}\n\nInput:\n{context}")
```

```python
hf_pipeline = HuggingFacePipeline(pipeline=generate_text)

llm_chain = LLMChain(llm=hf_pipeline, prompt=prompt, verbose=True)
llm_context_chain = LLMChain(llm=hf_pipeline, prompt=prompt_with_context, verbose=True)
```

```python
context = """George Washington (February 22, 1732[b] – December 14, 1799) was an American military officer, statesman,
and Founding Father who served as the first president of the United States from 1789 to 1797."""

instruction = "When was George Washington president?"

prediction = llm_context_chain.predict(instruction=instruction, context=context).lstrip()

print(prediction)
```

```python
context = """George Washington (February 22, 1732[b] – December 14, 1799) was an American military officer, statesman,
and Founding Father who served as the first president of the United States from 1789 to 1797."""

instruction = "When was George Washington president, and provide your rationale?"

prediction = llm_context_chain.predict(instruction=instruction, context=context).lstrip()

print(prediction)
```

```python
from langchain.prompts.few_shot import FewShotPromptTemplate
from langchain.prompts.prompt import PromptTemplate
```

```python
examples = [
{
'instruction':'What country does the Tigris run along?',
'input':'The name of the third river is Tigris; it runs along the east side of Assyria. The fourth river is the Euphrates.',
'response':'Assyria',
'rationale':
'''Read the instruction: What country does the Tigris run along?
Analyze the input to gather relevant information "The name of the third river is Tigris; it runs along the east side of Assyria. The fourth river is the Euphrates."
Identify the mention of the Tigris river in the input.
Note the statement that the Tigris river runs along the east side of Assyria.
Determine that the instruction is asking for the country that the Tigris river runs along.
Connect the information from the input to the instruction and conclude that "Assyria" is the country the Tigris river runs along.
Provide the response "Assyria" as it correctly identifies the country associated with the Tigris river.'''
},
{
'instruction':'Who went out from the presence of the Lord and lived in the land of Nod?',
'input':'So Cain went out from the presence of the Lord and lived in the land of Nod, east of Eden.',
'response':'Cain',
'rationale':
'''Read the instruction: "Who went out from the presence of the Lord and lived in the land of Nod?"
Analyze the input to gather relevant information: "So Cain went out from the presence of the Lord and lived in the land of Nod, east of Eden."
Identify the mention of Cain in the input.
Note the statement that Cain went out from the presence of the Lord and lived in the land of Nod.
Determine that the instruction is asking for the person who went out from the presence of the Lord and lived in the land of Nod.
Connect the information from the input to the instruction and conclude that "Cain" is the person who fits the criteria.
Provide the response "Cain" as it correctly identifies the person who went out from the presence of the Lord and lived in the land of Nod.'''
},
{
'instruction':'Who went out with his sons, his wife, and his sons’ wives?',
'input':'Noah went out along with his sons, his wife, and his sons’ wives.',
'response':'Noah',
'rationale':
'''Read the instruction: "Who went out with his sons, his wife, and his sons’ wives?"
Analyze the input to gather relevant information: "Noah went out along with his sons, his wife, and his sons’ wives."
Identify the mention of Noah in the input.
Note the statement that Noah went out with his sons, his wife, and his sons\' wives.
Determine that the instruction is asking for the person who went out with his sons, his wife, and his sons\' wives.
Connect the information from the input to the instruction and conclude that "Noah" is the person who fits the criteria.
Provide the response "Noah" as it correctly identifies the person who went out with his sons, his wife, and his sons\' wives.'''
}
]
```

```python
example_prompt = PromptTemplate(input_variables=['instruction', 'input', 'response', 'rationale'],
                                template="Question: {instruction}\nInput: {input}\nAnswer: {response}\nRationale: {rationale}")

print(example_prompt.format(**examples[2]))
```

```python
prompt = FewShotPromptTemplate(
    examples=examples, 
    example_prompt=example_prompt, 
    suffix='Question: {instruction}\nInput: {input}', 
    input_variables=['instruction','input']
)
```

```python
context = 'The sons of Gomer were Askenaz, Riphath, and Togarmah.'

prediction = llm_chain.predict(instruction=prompt.format(instruction='Who was the father of Askenaz?', input=context)).lstrip()

print(prediction)
```
